```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

# The given URL is set

```python
url = 'https://www.bandsintown.com/?came_from=257&hide_homepage=true&sort_by_filter=Number+of+RSVPs'
```

```python
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
soup
```

```python
soup.findAll('a')
```

```python
one_a_tag = soup.findAll('a')[55]
```

```python
link = one_a_tag['href']
link
```

```python
line_count = 1
for one_a_tag in soup.findAll('a'): 
    if line_count >= 55: 
        link = one_a_tag['href']
        download_url = 'https://www.bandsintown.com/'+ link
        urllib.request.urlretrieve(download_url,'./'+link[link.find('a')+1:]) 
        time.sleep(1)
    line_count +=1
```
